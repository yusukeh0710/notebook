## Color Histogram

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
import cv2

image = cv2.imread("sample1.jpg", 0)
plt.imshow(image, 'gray')
```

```python
plt.hist(image.ravel(),256,[0,256])
plt.show()
```

```python
image = cv2.imread("sample2.jpg", -1)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

```python
plt.figure(1)

plt.subplot(311)
red = image[:,:,0]
plt.hist(red.ravel(), 256, [0, 256], fc='r')

plt.subplot(312)
green = image[:,:,1]
plt.hist(green.ravel(), 256, [0, 256], fc='g')

plt.subplot(313)
blue = image[:,:,2]
plt.hist(blue.ravel(), 256, [0, 256], fc='b')

plt.show()
```
